## Scrapping all the Flag Images of all Countries

Importing required Libraries

```python
import requests 
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
# url of the website from which we will be scrapping.
url = 'https://www.worldatlas.com/countries'
```

```python
# Creating a response to get all the data from the url.
response = requests.get(url)
```

```python
# Using BeautifulSoup library to parse the data so we can extract the required information and image urls.
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Extracting Names of the countries.
name = soup.find_all('h3')
```

```python
# Creating a list of all country names.
country = []

for name in name:
    country.append(name.text)
```

```python
country
```

```python
# Extracting Country Codes.
code = soup.find_all('td', {'id' : 'item_iso3'})
```

```python
# Creating a list of all country codes.
country_code = []

for code in code:
    country_code.append(code.text)
```

```python
country_code
```

```python
# Extracting the image URLs.
image = soup.find_all('img')
```

```python
# Creating a list of all country flag image URLs.
country_flag = []

for flag in image[:-1]:
    country_flag.append(flag['src'])
```

```python
country_flag
```

```python
# Creating a dictionary so that we can merge all the lists and convert them into a DataFrame.
flag_dict = {
    'Country' : country,
    'Country Code' : country_code,
    'Country Flag' : country_flag
    
}
```

```python
#Creating a DataFrame.
flag_df = pd.DataFrame(flag_dict)
```

```python
flag_df
```

```python
# Converting the DataFrame into a csv file.
flag_df.to_csv('IDS_CountryFlag.csv', index=None)
```
